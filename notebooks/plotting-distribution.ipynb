```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
gauss = torch.distributions.Normal(torch.tensor([0.0]), torch.tensor([1.0]))
gauss2 = torch.distributions.Normal(torch.tensor([5.0]), torch.tensor([1.0]))
```

```python
gauss_samples = gauss.sample((10000,))
```

```python
_ = plt.hist(x=gauss_samples.numpy(),bins=100)
```

```python
sns.distplot(gauss_samples, kde=True)
```

```python
gauss_samples2 = gauss2.sample((10000,))
```

```python
print (torch.cat((gauss_samples,gauss_samples2)).shape)
```

```python
sns.distplot(torch.cat((gauss_samples,gauss_samples2)), kde=True)
```
